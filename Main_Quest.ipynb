```python
import random # random 모듈 호출

class Account:
  account_count = 0

  def __init__(self, name, money):
    self.bank = 'SC은행'
    self.name = name
    self.money = money
    self.account_number = self.make_account_number()
    self.deposit_count = 0 # 입금 횟수
    self.withdraw_count = 0
    self.deposit_list = []
    self.withdraw_list = []

    Account.account_count += 1 # 계좌 생성 개수 저장

  def make_account_number(self):
      a = random.randint(100, 999)
      b = random.randint(10, 99)
      c = random.randint(100000, 999999)
      return f'{a}-{b}-{c}'

  def get_account_num(self):
      return Account.account_count

  def deposit(self, depo):
      while True:
        try:
          a = depo

          if a > 0 and a >= 1:
            self.money += a # 입금한 돈과 잔액 합산
            self.deposit_count += 1 # 입금 횟수 증가
            self.deposit_list.append((a, self.money))
            if self.deposit_count % 5 == 0:
              ray = int(self.money * 0.01)
              print(f'{ray}의 이자가 입금되었습니다.')
              self.money += ray
              self.deposit_list.append((ray, self.money))
            break

          elif a < 0:
            print('입금은 최소 1원 이상만 가능합니다.')
            break

        except ValueError:
          print('숫자만 입력해 주세요.')

  def withdraw(self, draw):
      while True:
        try:
          b = draw

          if b > 0 and b <= self.money:
            self.money -= b
            self.withdraw_count += 1
            self.withdraw_list.append((b, self.money))
            break

          elif b == 0:
            print('출금하지 않습니다.')
            break

          elif b < 0:
            print('출금은 최소 1원 이상만 가능합니다.')

          elif b > self.money:
            print('계좌 잔고 이상으로 출금할 수 없습니다.')
            break

        except ValueError:
          print('숫자만 입력해 주세요')


  def display_info(self):
    format_money = f'{self.money:,}'
    print(f'은행이름: {self.bank}, 예금주: {self.name}, 계좌번호: {self.account_number}, 잔고: {format_money}원')

  def deposit_history(self):
    for i, (a, b) in enumerate(self.deposit_list):
      format_a = f'{a:>5}'
      format_b = f'{b:>5}'
      print(f'{i + 1:>2}회     {self.name}님의 입금 내역: {format_a}원     잔액: {format_b}원')

  def withdraw_history(self):
    for i, (a, b) in enumerate(self.withdraw_list):
      format_a = f'{a:>5}'
      format_b = f'{b:>5}'
      print(f'{i + 1:>2}회     {self.name}님의 출금 내역: {format_a}원     잔액: {format_b}원')

# 3개 이상 인스턴스 생성
first_account = Account('luna', 1000)
second_account = Account('lena', 500000)
third_account = Account('rume', 1500000)

#잔고 100만원 이상 고객 정보만 출력
accounts = [first_account, second_account, third_account]

def whale(accounts):
  total = first_account.get_account_num()
  print('생성된 계좌의 총 개수:', total)
  for account in accounts:
    if account.money >= 1000000:
      account.display_info()

(whale(accounts))
print()

# 입금, 출금 테스트
my_account = Account('quest', 1000)
for i in range(10):
  my_account.deposit(1000)
my_account.deposit(-1000)
my_account.withdraw(50000)
my_account.withdraw(500)
print()

# 입금, 출금 내역 출력
my_account.deposit_history()
print('-' * 55)
my_account.withdraw_history()
```
